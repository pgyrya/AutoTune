```python
from AutoTune import StepSearchCV
```

```python
# prepare a toy dataset to play with 

import numpy as np
X = np.random.normal(scale = 1, size = 100000).reshape(5000,20) #500 observations and 20 variables
y = X[:,0] + X[:,1] + X[:,0]* X[:,1]
X.shape, y.shape
```

```python
from sklearn.neural_network import MLPRegressor
searcher = StepSearchCV(
    estimator        = MLPRegressor(),
    parameters_start = {'learning_rate_init': 0.1, 'batch_size': 64,'hidden_layer_sizes': (4,1)},  
    steps            = 1000,
    step_veto_rule   = lambda x,y: y<=0 or ((x >=0.95*y) and (x <=1.05*y)),
    save_location    = 'cv_results.csv'
)
```

```python
searcher.fit(X,y)
```

```python
# We could continue tuning process where we left off by requesting more steps
searcher.fit(X, y, steps_completed = 1000, more_steps = 1000)
```
